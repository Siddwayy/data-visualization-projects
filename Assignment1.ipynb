```python
import pandas as pd
import matplotlib.pyplot as plt

# 1. Load Data
df = pd.read_csv('river-plastics.csv')

# 2. Prepare Data
# We use the exact names from your screenshot
x_data = df['Coast length'] / df['Area']
c_data = df['Rainfall']
y_data = df.iloc[:, -1] # Selects the last column (Percentage) automatically

# 3. Create Plot
fig, ax = plt.subplots(figsize=(10, 6))
scatter = ax.scatter(x_data, y_data, c=c_data, cmap='viridis')

# 4. Configure Axes (Log Scale)
ax.set_xscale('log')
ax.set_yscale('log')

# 5. Add Labels & Title
ax.set_title('Coastal Length, Area, and Rainfall Effect on Plastic Emission')
ax.set_xlabel('Ratio of Coastal Length to Area')
ax.set_ylabel('Plastic Waste Emitted to Ocean (%)')

# 6. Add Colorbar
cbar = fig.colorbar(scatter, ax=ax)
cbar.set_label('Annual Rainfall (mm)')

# 7. Apply Assignment Colors ('#051B2C')
dark_blue = '#051B2C'

# Color the text
ax.title.set_color(dark_blue)
ax.xaxis.label.set_color(dark_blue)
ax.yaxis.label.set_color(dark_blue)
ax.tick_params(colors=dark_blue)

# Color the borders (spines)
for spine in ax.spines.values():
    spine.set_edgecolor(dark_blue)

# Color the colorbar
cbar.set_label('Annual Rainfall (mm)', color=dark_blue)
cbar.ax.yaxis.set_tick_params(color=dark_blue, labelcolor=dark_blue)
cbar.outline.set_edgecolor(dark_blue)

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# --- 1. Load the Data ---
# REPLACE 'sleep_data.csv' with the exact name of your file!
# Based on your screenshot, the headers in the CSV should be:
# 'Age group', 'Recommended lower', 'Recommended upper', etc.
df = pd.read_csv('sleep.csv')

# --- 2. Setup Plotting Variables ---
# Assignment colors
dark_color = '#051B2C'   # Dark Blue (Lower bound)
light_color = '#4789BE'  # Lighter Blue (Upper bound)

# Create figure
fig, ax = plt.subplots(figsize=(8, 6))

# --- 3. Plot the Data ---
# We iterate through the dataframe rows to draw each "dumbbell"
for i in range(len(df)):
    # Extract values from the CSV columns
    # We use the exact headers seen in your screenshot
    lower = df.loc[i, 'Recommended lower']
    upper = df.loc[i, 'Recommended upper']

    # Draw the Line (connecting the dots)
    ax.plot([lower, upper], [i, i], color='black', zorder=1)

    # Draw the Dots
    # Lower dot (Dark Color)
    ax.scatter(lower, i, color=dark_color, s=50, zorder=2)
    # Upper dot (Light Color)
    ax.scatter(upper, i, color=light_color, s=50, zorder=2, edgecolors='black')

# --- 4. Customize Axes ---

# Set Y-axis labels using the 'Age group' column
ax.set_yticks(range(len(df)))
ax.set_yticklabels(df['Age group'])

# Invert Y-axis so the first row (Newborns) is at the top
ax.invert_yaxis()

# Move X-axis to the TOP
ax.xaxis.tick_top()
ax.xaxis.set_label_position('top')

# Set specific X-axis ticks as shown in the target image
ax.set_xticks([0, 6, 8, 10, 12, 14, 18, 24])

# Remove Spines (borders) on Left, Right, and Bottom
ax.spines['left'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['bottom'].set_visible(False)

# Remove Y-axis ticks (keep labels, remove the little lines)
ax.tick_params(axis='y', length=0)

# --- 5. Add Title ---
plt.title("National Sleep Foundation's recommended hours of sleep", pad=20)

# Show the plot
plt.show()
```
