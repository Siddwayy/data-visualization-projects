Question 1 - Scatter Plot

```python
import pandas as pd
import matplotlib.pyplot as plt

# load the data
df = pd.read_csv('river-plastics.csv')

# calculate the x-axis ratio
df['coast_area_ratio'] = df['Coast length'] / df['Area']

# filter data so log scale works (must be positive)
plot_data = df[(df['coast_area_ratio'] > 0) & (df['Ratio ME/MPW'] > 0)]

# set figure size
plt.figure(figsize=(10, 6))

# create the scatter plot using rainfall for color
sc = plt.scatter(plot_data['coast_area_ratio'], plot_data['Ratio ME/MPW'],
                 c=plot_data['Rainfall'], cmap='viridis')

# use log scale for both axes
plt.xscale('log')
plt.yscale('log')

# set text color
text_color = '#051B2C'

# add labels and title with the specific color
plt.xlabel('Ratio of Coastal Length to Area', color=text_color)
plt.ylabel('Plastic waste emitted in oceans(% of total plastic waste)', color=text_color)
plt.title('Drivers of Plastic Pollution in the Oceans', color=text_color)

# add the colorbar
plt.colorbar(sc, label='Rainfall [mm/year]')

# show the final plot
plt.show()
```

Question 2 – Connected Dot Plot

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# load the data
df = pd.read_csv('sleep.csv')

# sort by the lower recommended value
df = df.sort_values(by='Recommended lower', ascending=False)

# create the figure
fig, ax = plt.subplots(figsize=(8, 6))

# loop through each row to draw the plot
for i, row in df.iterrows():
    # draw the line connecting the two dots
    ax.plot(row[['Recommended lower', 'Recommended upper']], [i, i], linestyle='-', color='black')
    # plot the blue dot (upper value)
    ax.plot(row['Recommended upper'], i, marker='o', markersize=8, color='#4789be')
    # plot the dark dot (lower value)
    ax.plot(row['Recommended lower'], i, marker='o', markersize=8, color='#051b2c')

# set y-axis ticks and labels
ax.set_yticks(np.arange(len(df)))
ax.set_yticklabels(df['Age group'])

# flip y-axis so the list starts at the top
ax.invert_yaxis()

# put x-axis labels on top
ax.xaxis.tick_top()

# set custom x-axis numbers
ax.set_xticks([0, 6, 8, 10, 12, 14, 18, 24])
ax.set_xticklabels([0, 6, 8, 10, 12, 14, 18, 24], ha='right')

# remove the border lines
ax.spines['left'].set_visible(False)
ax.spines['bottom'].set_visible(False)
ax.spines['right'].set_visible(False)

# add title
ax.set_title("National Sleep Foundation's recommended hours of sleep", loc='left')

# show the plot
plt.show()
```
