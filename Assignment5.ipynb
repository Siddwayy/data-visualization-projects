### Question 1 – Interactive Heat Map (Average Carat)


```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
from ipywidgets import widgets, interactive

diamonds_df = pd.read_csv(r'diamonds.csv')

def plot_heatmap(origin_filter):
    if origin_filter != 'All':
        filtered_data = diamonds_df[diamonds_df['type'] == origin_filter.lower()]
    else:
        filtered_data = diamonds_df

    pivot_table = filtered_data.pivot_table(values='carat', index='colour', columns='clarity', aggfunc='mean')

    clarity_ordered = ['SI2', 'SI1', 'VS2', 'VS1', 'VVS2', 'VVS1', 'IF']
    color_ordered = ['D', 'E', 'F', 'G', 'H', 'I', 'J']

    pivot_table = pivot_table.reindex(index=color_ordered, columns=clarity_ordered)

    plt.figure(figsize=(10, 8))
    sns.heatmap(pivot_table, cmap="mako", annot=False)
    plt.title('Average Carat Weight of Diamonds by Clarity and Color')
    plt.xlabel('Clarity')
    plt.ylabel('Color')
    plt.show()

dropdown = widgets.Dropdown(options=['All', 'Natural', 'Lab'])

interactive_widget = interactive(plot_heatmap, origin_filter=dropdown)

interactive_widget
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import ipywidgets as widgets


names = pd.read_csv('names.csv')
names['name_lower'] = names['name'].str.lower()
COLORS = ['#206095', '#004662', '#A8BD3A', '#27A0CC', '#118C7B', '#F66068']

def render(sex: str, highlight_text: str):

    tokens = str(highlight_text or '').lower().split()
    requested = list(dict.fromkeys(tokens))[:6]

    df_sex = names[names['sex'] == sex]
    top_100 = df_sex[df_sex['year'] == 2020].nsmallest(100, 'rank')['name_lower']
    df_plot = df_sex[df_sex['name_lower'].isin(top_100)]

    fig, ax = plt.subplots(figsize=(14, 6))

    for _, d in df_plot.groupby('name_lower'):
        ax.plot(d['year'], d['rank'], color='#BDBDBD', alpha=0.35, linewidth=1)


    highlight_present = [nm for nm in requested if nm in top_100.values]
    for i, nm in enumerate(highlight_present):
        d = df_plot[df_plot['name_lower'] == nm].sort_values('year')
        ax.plot(d['year'], d['rank'], color=COLORS[i], linewidth=2.5, label=nm.capitalize())
        ax.scatter(d['year'], d['rank'], color=COLORS[i], s=16, zorder=3)

    if highlight_present:
        ax.legend(frameon=False, loc='upper right')

    ax.set_title(f'Top baby {sex} names by popularity rank (England and Wales)')
    ax.set(xlabel='', ylabel='', xlim=(1996, 2020), ylim=(1000, 1))
    ax.set_xticks(range(1996, 2021, 2))
    ax.set_yticks([1] + list(range(100, 1001, 100)))
    ax.grid(axis='y', linestyle='-', linewidth=0.6, alpha=0.35)

    for spine in ['left', 'right', 'top']:
        ax.spines[spine].set_visible(False)

    return fig


toggle = widgets.ToggleButtons(options=['boy', 'girl'], value='boy', description='Names:')
text_box = widgets.Text(value='', description='Highlight:', placeholder='Type names separated by spaces')
out = widgets.Output()

def update(*args):
    out.clear_output(wait=True)
    with out:
        plt.show(render(toggle.value, text_box.value))

toggle.observe(update, names='value')
text_box.observe(update, names='value')

update()

widgets.VBox([toggle, text_box, out])
```
